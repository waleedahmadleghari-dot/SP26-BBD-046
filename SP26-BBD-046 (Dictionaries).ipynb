```python
orders = [
    {"product": "Shirt", "category": "Clothing"},
    {"product": "Pants", "category": "Clothing"},
    {"product": "Phone", "category": "Electronics"},
    {"product": "Laptop", "category": "Electronics"},
    {"product": "TV", "category": "Electronics"}
]

category_count = {}

for order in orders:
    category = order["category"]
    if category in category_count:
        category_count[category] += 1
    else:
        category_count[category] = 1
print (category_count)
most_popular = max(category_count, key=category_count.get)

print("Most popular category:", most_popular)
```

```python
inventory = {}
inventory["Apple"] = {"quantity": 10, "price": 50}
inventory["Banana"] = {"quantity": 3, "price": 30}
inventory["Apple"]["quantity"] = 15
total = 0
for item in inventory.values():
    total += item["quantity"] * item["price"]
print("Total Value:", total)
for name, item in inventory.items():
    if item["quantity"] < 5:
        print(name, "is low on stock")
```

```python
students = {
    "Waleed": {"Math": 50, "English": 90},
    "Hani": {"Math": 92, "English": 88},
    "Ahmed": {"Math": 70, "English": 70}
}
for name, subjects in students.items():
    avg = sum(subjects.values()) / len(subjects)
    print(name, "Average:", avg)
grades = {}
for name, subjects in students.items():
    grades[name] = subjects["Math"]
highest = max(grades, key=grades.get)
lowest = min(grades, key=grades.get)
print("Highest in Math:", highest)
print("Lowest in Math:", lowest)
sorted_students = sorted(
    students.items(),
    key=lambda x: sum(x[1].values()) / len(x[1]),
    reverse=True
)
print("Sorted by average:")
for student in sorted_students:
    print(student[0])
```

```python
store = {
    "Laptop": {"price": 1000, "quantity": 5},
    "Phone": {"price": 500, "quantity": 10}
}
cart = {}
product = "Laptop"
qty = 1
if product in store and store[product]["quantity"] >= qty:
    cart[product] = cart.get(product, 0) + qty
product = "Phone"
qty = 2
if product in store and store[product]["quantity"] >= qty:
    cart[product] = cart.get(product, 0) + qty
if "Phone" in cart:
    del cart["Phone"]
total = 0
for product, qty in cart.items():
    total += store[product]["price"] * qty
print("Total:", total)
discount = 0.1 
tax = 0.05  
final_total = total - (total * discount)
final_total = final_total + (final_total * tax)
print("Final Bill:", final_total)
```
