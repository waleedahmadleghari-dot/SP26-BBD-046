```python
def calculate_average(a, b, c):
    return (a + b + c) / 3

print(calculate_average(10, 20, 30))
```

```python
def calculate_discount(price, discount=10):
    return price - (price * discount / 100)

print(calculate_discount(1000))        
print(calculate_discount(1000, 20))    
```

```python
def find_max(*numbers):
    return max(numbers)

print(find_max(3, 7, 2, 9, 5))
```

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
def apply_function(func, numbers):
    result = []
    for n in numbers:
        result.append(func(n))
    return result

def square(x):
    return x * x

print(apply_function(square, [1, 2, 3, 4]))
```

```python
def apply_operation(operation, numbers):
    result = []
    for num in numbers:
        result.append(operation(num))
    return result
def add_ten(x):
    return x + 10
numbers = [1, 20, 83]
output = apply_operation(add_ten, numbers)
print("After adding 10:", output)
```
