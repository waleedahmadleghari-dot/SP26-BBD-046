```python
secret = 25
attempts = 0
guess = int(input("Guess the number: "))
while guess != secret:
    attempts += 1
    
    if guess > secret:
        print("Too high")
    else:
        print("Too low")
    
    guess = int(input("Guess again: "))

attempts += 1
print("Correct! Attempts:", attempts)
```
